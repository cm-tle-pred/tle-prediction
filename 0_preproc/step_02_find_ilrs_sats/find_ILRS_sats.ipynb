```python
import os
import pandas as pd
pd.set_option('display.max_columns', 500)

# Get GP data
df = pd.read_csv(os.environ['my_home_path'] + "/data/space-track-gp/gp_20210129.csv.gz")
```

```python
# Get the ILRS satellites
pres_html = pd.readï¼¿html('https://ilrs.gsfc.nasa.gov/missions/satellite_missions/current_missions/')
past_html = pd.read_html('https://ilrs.gsfc.nasa.gov/missions/satellite_missions/past_missions/index.html')

pres_df = pres_html[0].iloc[:,0:4]
past_df = past_html[0]

pres_df.columns = ['name', 'cospar_id', 'sic_code', 'norad_id']
pres_df['last_update'] = 'present'

past_df.columns = ['name', 'cospar_id', 'sic_code', 'norad_id', '0', '1', '2','3','4','last_update']
past_df = past_df[['name', 'cospar_id', 'sic_code', 'norad_id','last_update']]

ilrs_sat_df = pd.concat([pres_df,past_df])
print(len(ilrs_sat_df))
```

```python
# Merge the ILRS sats with GP data
df = df.merge(ilrs_sat_df, how='left', left_on='NORAD_CAT_ID', right_on='norad_id')
df = df[df.name.notnull()].sort_values('name')
print(len(df[df.name.notnull()]))
```

```python
# Get only LEO satellites
leo_df = df[(df['MEAN_MOTION']>11.25) & (df['ECCENTRICITY']<0.25)]
print(len(leo_df))
```

```python
# Get only ones that are still being tracked
present_leo_df = leo_df[leo_df.last_update == 'present']
print(len(present_leo_df))
```
